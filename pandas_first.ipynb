# Pandas Series
A Pandas Series is a one-dimensional labeled array capable of holding data of any type (integers, strings, floating-point numbers, Python objects, etc.). You can think of it as a single column in an Excel spreadsheet or a database table, where each value is paired with a specific row label called an index

## Missing Data

Finding Missing Data

```python
data = {
    'A': [1,2, np.nan, 4,5],
    'B': [1,2,3,4,5 ],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5],
    'E':[1,2,3,4,5]
}
df_find_m =pd.DataFrame(data)
df_find_m
```

```python
df_find_m.isna()
```

```python
df_find_m.isna().sum()
```

```python
df_find_m.isna().any()
```

Removing the Missing Data

```python
df_find_m.dropna()
```

```python
df_find_m.dropna(thresh = 4)
```

Filling the missing value


```python
df_find_m
```

```python
df_find_m.fillna(0)
```

```python
values = {"A":0,"B":100,"C":300,"D":400,"E":500}
df_find_m.fillna(value=values)
```

```python
df_find_m.fillna(df_find_m.mean())
```

# Merging Joining Concatenation of 2 DataFrame Joining of 2 DataFrames

Merging to dataframes


```python
# DataFrame 1 : employees 
employees = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['John','Anna','Peter','Linda','Bob'],
    'department':['HR','IT','Finance','IT','AI']
})
# DataFrame 2: salaries
salaries = pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salary':[60000,80000,65000,70000,90000],
    'bonus':[5000,10000,7000,8000,12000]

})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries )
```

```python
pd.merge(employees,salaries, on = 'employee_id',how = 'inner')
```

```python
pd.merge(employees,salaries,on = 'employee_id',how = 'outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how = 'left')
```

```python
pd.merge(employees,salaries,on = 'employee_id',how = 'right')
```
